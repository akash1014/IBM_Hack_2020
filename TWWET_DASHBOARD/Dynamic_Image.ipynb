```python
import matplotlib.pyplot as plt
import sched, time

s = sched.scheduler(time.time, time.sleep)
a=0
def call_files():
    global a
    print("in call files"+str(a)+"::")
    read_file="sample"+str(a)+".txt"
    pulldata=open(read_file,"r").read()

    if(a+1==10):
        a=0
    else:
        a=a+1
    return pulldata


def do_something(sc):
    fig=plt.figure()
    ax12=fig.add_subplot(1,1,1)
    pulled=call_files()
    plt.title("Doing stuff...")

    dataArray=pulled.split('\n')
    xar=[]
    yar=[]
    for eachline in dataArray:
        if len(eachline)>1:
            x,y=eachline.split(',')
            xar.append(int(x))
            yar.append(int(y))
    ax12.clear()
    ax12.plot(xar,yar)
    print(xar)
    plt.show()
    # do your stuff

    s.enter(10, 1, do_something, (sc,))

s.enter(10, 1, do_something, (s,))
s.run()
```
